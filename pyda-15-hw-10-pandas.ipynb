```python
data_power = pd.read_csv('power.csv')
data_power
```

```python
data_power.info()
```

```python
# находим как называются Латвия, Литва и Эстония в  датафрейме

data_power[ data_power['country'].str.contains('est', case=False) ]['country'].unique()
```

```python
filtered_data = (data_power[((data_power['country'] == 'Latvia') |
                             (data_power['country'] == 'Lithuania') | 
                             (data_power['country'] == 'Estonia')) 
                            & 
                            ((data_power['category'] == 4) |
                             (data_power['category'] == 12) |
                             (data_power['category'] == 21)) 
                            &
                            (2005 <= data_power['year']) & (data_power['year']<= 2010)])
```
